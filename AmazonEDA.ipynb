***1.what is the avarage rating count for each product category?***

```python
avg_rating_by_category = df.groupby('category')['rating'].mean()
print(avg_rating_by_category)
```

***2.What are the top rating_count _products by category?***

```python
top_rating_count_products = df.loc[df.groupby('category')['rating_count'].idxmax()]
print(top_rating_count_products)
```

***3.what is the distribution of discounted price vs actual price?***

```python
df.columns
```

```python
df['discounted_price'].unique()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df['actual_price'].dtype
```

***4.how does the avarage discount price vary accross categories?***

```python
df.columns
```

```python
df['discount_percentage'].unique()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
df['discount_percentage'].dtype
```

```python
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean()
print(avg_discount_by_category)
```

```python
plt.figure(figsize=(20, 11))
sns.barplot(x=avg_discount_by_category.index, y=avg_discount_by_category.values)
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***5.what are the most popular product names?***

```python
popular_products = df.groupby('product_name')['product_name'].count().sort_values(ascending=False)
print(popular_products.head(10))
```

***6.what are the most popular product keywords?***

```python
from collections import Counter
all_product_names = ' '.join(df['product_name'].astype(str).tolist())
keywords = all_product_names.lower().split()
keyword_counts = Counter(keywords)
most_common_keywords = keyword_counts.most_common(10)
print(most_common_keywords)
```

```python
keywords, counts = zip(*most_common_keywords)

plt.figure(figsize=(12,11))
sns.scatterplot(x=keywords, y=counts,alpha=1)
plt.title('Top 10 Most Popular Product Keywords')
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
