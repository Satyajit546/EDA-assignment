***2.What is the distribution of popularity among  the track in the data set?visualize it using a histogram.***

```python
df['Popularity'].unique()
```

```python
df['Popularity'].value_counts()
```

```python
distribution_of_popularity = df.groupby('Track Name')['Popularity'].describe()
print(distribution_of_popularity)
```

```python
#Visualize the distribution with histogram:
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=30,kde=True,)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

***5.What are the top 5 most poupular tracks in the dataset? Provide the artist name and track name for each.***

```python
top_5_tracks = df.sort_values(by=['Popularity'], ascending=False).head(5)
print(top_5_tracks[['Artist', 'Track Name']])
```

```python
#Lets visualize the plot with barplot:
plt.figure(figsize=(15, 10))  # Adjust figure size as needed
sns.barplot(x='Popularity', y='Track Name', hue='Artist', data=top_5_tracks, dodge=False, palette='viridis')
plt.title('Top 5 Most Popular Tracks')
plt.xlabel('Popularity')
plt.ylabel('Track Name')
plt.legend(title='Artist', loc='upper right', bbox_to_anchor=(1.2, 1),fontsize=12)
plt.tight_layout()
plt.show()
```

***6.Among the top 5 most popular artist,Which artist has the highest popularity on avarage? Calculate and display the avarage popularity for each artist.***

```python
#Find the Top 5 artist:
top_5_artists = df['Artist'].value_counts().head(5).index.tolist()
print(top_5_artists)
```

```python
#Avarage polpularity:
average_popularity = df[df['Artist'].isin(top_5_artists)].groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
print(average_popularity)
```

```python
# visualize with line plot:
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_popularity.index, y=average_popularity.values, marker='o', color='blue')
plt.title('Average Popularity of Top 5 Artists')
plt.xlabel('Artist')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

***8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.***

```python
plt.figure(figsize=(25, 15))
sns.pairplot(data=df,vars=['Popularity', 'Duration (ms)'],hue='Artist',diag_kind='kde')
plt.legend(title='Artist', loc='right', bbox_to_anchor=(5,.5),fontsize=4)
plt.tight_layout()
plt.show()
```

***9.Does the duration of tracks vary significantly accross different artists? Explore this visually using box plot or violin plot.***

```python
# Visualize the duration of tracks vary significantly accross different artists:

plt.figure(figsize=(30, 10))
sns.violinplot(x='Artist', y='Duration (ms)', data=df,palette="pastel")
plt.title('Duration of Tracks Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```

Insight:

This helps to observe how the duration of tracks varies between artists. The distribution shapes and central tendencies for each artist's tracks duration might indicate differences in their musical styles or preferences.

***10.How does the distribution of track popularity vary for different artists? visualize this using a swarm plot or a violin plot.***

```python
plt.figure(figsize=(15, 10))
sns.swarmplot(x='Artist', y='Popularity', data=df, palette='viridis')
plt.title('Distribution of Track Popularity for Different Artists (Swarm Plot)')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Individual Data Points:

Swarm plots are especially useful for viewing individual data points for each artist. This can help identify clusters of tracks within a certain popularity range or spot potential outliers more clearly.

Overlapping Points:

Overlapping points in a swarm plot can indicate areas of higher density, meaning more tracks fall within that particular popularity range.
